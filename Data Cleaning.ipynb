HANDLING MISSING DATA:
There are a few ways to filter out missing data.
a) Filling the missing data
b) Dropping the missing data colums/rows

While you always have the option to do it by hand using pandas.isnull and boolean indexing,
the dropna and fillna can be helpful.

FILLING THE MISSING DATA:
â€¢ Rather than filtering out missing data (and potentially discarding other data along with
it), you may want to fill in the â€œholesâ€ in any number of ways. For most purposes, the
fillna method is the workhorse function to use. Calling fillna with aconstant replaces
missing values with that value.
DROPPING THE MISSING DATA ROWS/COLUMNS:
â€¢ With DataFrame objects, things are a bit more complex. You may want to drop rows or
columns that are all NA or only those containing any NAs. dropna by default drops any
row containing a missing value.

```python
# fillna method , it fills all missing columns with 0
abc=df.fillna(0)
abc
```

```python
#FILLNA WITH FFILL: fill with preceeding value
abc=df.fillna(method="ffill")
abc
```

```python
#FILLNA WITH MEAN:
abc=df.fillna(df.mean())
abc
```

```python
a = df.fillna(0) #fill all with 0 or any other value
b = df.fillna(method="ffill") #fill with the preceeding value
c= df.fillna(df.mean) #fill with the mean value
```

```python
#DROPNA FOR ROWS:
xyz=df.dropna() #for droping rows
xyz
```

```python
#REPLACE UNEXPECTED VALUES:
df.loc[df["Parch"].isin(["-999"])]="NO"
df
```
